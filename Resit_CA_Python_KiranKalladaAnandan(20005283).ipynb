```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
# import re
```

```python
url="https://www.coverscart.com/iPhone-13-Mini/iPhone-13-Mini-Glass"
```

```python
r=requests.get(url)
print(r)
```

```python
soup=BeautifulSoup(r.text)
```

```python
print(soup)
```

```python
prod_name=[]
prod_discount_price=[]
prod_mrp=[]
product_images=[]
prod_descp=[]
```

```python
products=soup.find_all("div",class_='product-thumb')
for product in products:

  # Extracting name of the product
  name=product.find("div",class_='name').get_text(strip=True)
  prod_name.append(name)

  # Extracting Price
  price_text=product.find('div',class_='price').get_text(strip=True)
  prices=price_text.split('Rs.')
  discounted_price = int(prices[1].replace(',', '').split('.')[0])
  mrp = int(prices[2].replace(',', '').split('.')[0])
  prod_discount_price.append(discounted_price)
  prod_mrp.append(mrp)

  images=product.find('img')['src']
  product_images.append(images)

  description_tag=product.find("div",class_='description')
  description=description_tag.get_text(strip=True)if description_tag else None
  prod_descp.append(description)
print(len(prod_name))
print(len(prod_discount_price))
print(len(prod_mrp))
print(len(product_images))
print(len(prod_descp))



```

```python
min_length = min(len(prod_name), len(prod_discount_price), len(prod_mrp),
                 len(product_images), len(prod_descp))
prod_name = prod_name[:min_length]
prod_discount_price = prod_discount_price[:min_length]
prod_mrp = prod_mrp[:min_length]
product_images = product_images[:min_length]
prod_descp = prod_descp[:min_length]
```

```python
data={'Product Name':prod_name,
      'Discounted Price':prod_discount_price,
      'MRP':prod_mrp,
      'Images':product_images,
      'Description': prod_descp
}

df=pd.DataFrame(data)
```

```python
df
```

```python
df.info()
```

```python
def classify_price(price):
  if price==399:
    return 'Cheap'
  elif price<=499:
    return 'Average'
  elif price==599:
    return 'Expensive'
  else:
    return 'Not in List'
```

```python
df_classified=df.copy()
df_classified['Price Category']=df_classified['Discounted Price'].apply(classify_price)
df_classified['Discount Percentage']=(((df_classified['MRP']-df_classified['Discounted Price'])/df_classified['MRP'])*100).round()
```

```python
df_classified['Case Type']=df_classified['Product Name'].str.extract('(Glass Back Case)',expand=False)
```

```python
df_classified['Product Name']=df_classified['Product Name'].str.replace('Glass Back Case for iPhone 13 Mini','',regex=False)
```
